```python
import pandas as pd
import numpy as np


df = pd.read_csv('water-treatment.csv', header=0)


df.columns[0]
print(np.nan)
del df[df.columns[0]]
```

```python
df.replace({'\?':np.nan}, regex=True, inplace=True)
```

```python
df
```

```python
df.shape[0]
```

```python
drop_columns = []
num_col = len(df.columns)

for column in df.columns:
    
    count = (len(df[column].value_counts(dropna=True)))
    if (count/num_col) < 0.85:
        drop_columns.append(column)

for column in drop_columns:
    del df[column]


for column in df.columns:
    df[column].fillna(df[column].median())
    
```

```python
# try this KMeans model

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=0).fit(df)

print(kmeans)

```
