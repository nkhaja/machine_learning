```python
#import libraries
import pandas as pd
import numpy as np
from sklearn import datasets

iris = datasets.load_iris()
all_data = datasets.load_breast_cancer()

numbers = all_data['data']
target = all_data['target']
target_names = all_data['target_names']
feature_names = all_data['feature_names']

df = pd.DataFrame(data=numbers, columns=feature_names)
df['target'] = target

#Check how to drop the first label


```

```python
# Fill in missing values here

for column in df.columns:
    df[column].fillna(df[column].median())
    
```

```python
df
```

```python
import matplotlib.pyplot as plt
# Visualizing the data

for column in df.columns:
    plt.hist(df[column][df["target"] == 1], bins = 5)
    plt.hist(df[column][df["target"] == 0], bins = 5)
    plt.title(column)
    plt.show()

    
# for i in range len(df.columns)-1:
#     plt.figure(i)
#     plt.hist(df[df.columns[column]][df["Cancer"] == 0], bins = 5)

# plt.hist(df["mean texture"][df["target"] == 1], bins = 5)
    


```

```python
# do the train test split

from sklearn.model_selection import train_test_split

train, test = train_test_split(df, train_size = 0.7)

```

```python
# Build the model

from sklearn.naive_bayes import GaussianNB

target_features_train = train[["mean texture", "mean compactness", "worst perimeter"]].values
cancer_target_train = train["target"].values

gnb = GaussianNB().fit(target_features_train, cancer_target_train)
print(gnb.score(target_features_train, cancer_target_train))

```

```python
from sklearn.metrics import f1_score


# check te prediction against the test set 
target_features_test = test[["mean texture", "mean compactness", "worst perimeter"]].values
cancer_target_test = test["target"].values

# Decision tree prediction
predict = gnb.predict(target_features_test)
score = f1_score(cancer_target_test, predict)
print(score)
```
