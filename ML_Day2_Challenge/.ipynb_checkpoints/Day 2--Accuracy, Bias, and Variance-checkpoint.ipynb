Both of these doctors believe the're accurate.  In their own way, both of them are right!  Dr. Patrick Harris focused on **_recall_**, whereas Doctor Strange focused on **_precision_**.  Let's define these terms before we go any further.  

**Precision:** The percentage of correct predictions out of all predictions made.  

The formula for precision is:  True Positives / (True Positives + False Positives).

**Recall:** The percentage of correct predictions of a class, out of all actual instances of that class in a population.  

The formula for Recall is:  True Positives / (True Positives + False Negatives).

As a reminder, let's quickly review some terms.  These are easiest to explain using examples, so we'll stick with our medical theme:

**_True Positive:_** When the patient has a disease, and the test comes back positive.  
**_True Negative:_** When the patient doesn't have the disease, and the test comes back negative.  
**_False Positive:_** When the patient doesn't have the disease, but the test comes back positive. 
**_False Negative:_** When the patient has a disease, but the test comes back negative.

With this information, we can calculate precision, recall, and most importantly, our **_accuracy_**.  The formula for accuracy is pretty intuitive:

(True Positive + True Negative) / (True Positive + True Negative + False Positive + False Negative)

This means our accuracy is all of our correct guesses (or correct non-guesses, in the case of True Negatives), divided by all items.

This time around, we'll also calculate the F1 Score, which is one of the more commonly used accuracy measures in machine learning.  F1 Scores are usually a little bit more trustworthy than basic accuracy, since it is computed using precision and recall.  The formula for calculating F1 Score is:

F1 = 2 * (precision * recall) / (precision + recall)

<center><h2>Part 3: Bias, Variance, and Model Evaluation</h2></center>

For the last part of this exercise, we'll examine **_bias_** and **_variance_** in models.  Before we do, let's review these terms and what they mean in the world of machine learning.  

<h4>Bias:</h4> Error from incorrect assumptions in your model.  

<h4>Variance:</h4> Error from sensitivity to small fluctuations ("noise") in your data set.  

<img src="bias_variance_examples.png" height=400 width=400>

Let's use this chart as a way to check our understanding of these terms.  A model with **_high bias_** ignores the data in misses our actual target.  In machine learning, bias happens when we **_underfit_** our model.  This usually means that our model is too simple.  Conversely, variance (also known as sensitivity) is when we **_overfit_** our model. This usually means that our model has overtrained on our training data.  We'll discuss how this applies to supervised learning in a bit more detail, and then we'll see examples of an underfit and an overfit model, so that we can see what they look like.  

 
 
<center><h2>Model Fitting in Supervised Learning</h2></center>


In order to fully grasp what we're doing here, it's worth reviewing the entire goal of supervised learning.  We have a set of inputs, *X*, and we want to figure out a function that will spit out the correct output for each, *Y*.  Technically, there exists some perfect version of this function--in data science, we refer to this as the **_Ground Truth_**.  In most cases, you'll never know the actual ground truth, but that doesn't mean we can't approximate it and get close to it!

Ground Truth:  F(X) = Y 
(X is our input, Y is the correct label).  

So how much error is okay?  Where do we draw the line between an acceptable and unacceptable approximation?  How close is "close enough?"  This is a tough question to answer, because it partially depends on the problem you're trying to solve, the quality of your data set, etc.  However, there are certain metrics are generally accepted as good accuracy measures, such as F1 Score.  We'll talk in much more depth about model fitting and error metrics when we get into linear regressions other supervised learning algorithms.  

For now, let's close out by running some code and seeing if you can visually identify which model is underfit, which model is overfit, and which model is juuuuuust right.  

The following code is pulled straight from sklearn's (amazingly detailed!) documentation. Play around in the code and see if you can decipher what they're doing to create these models and graph their visualizations--this is a great piece of code to help you get a better feel for how to use numpy, sklearn, and matplotlib.  When you run the code, it will spit out visualizations of three different models of the same data.  Read the graphs and look for the telltale signs of underfitting and overfitting.  

As a final note, remember that figuring out if the model is overfit or underfit is easy in this example, because we can see the actual line of best fit.  In real world examples, this won't be the case--you'll be trying to approximate this, and it will be up to you to think scientifically and decide if your model is a good approximation of the underlying function that you're trying to discover--a much harder task than here!

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

np.random.seed(0)

n_samples = 30
degrees = [1, 4, 15]

true_fun = lambda X: np.cos(1.5 * np.pi * X)
X = np.sort(np.random.rand(n_samples))
y = true_fun(X) + np.random.randn(n_samples) * 0.1

plt.figure(figsize=(14, 5))
for i in range(len(degrees)):
    ax = plt.subplot(1, len(degrees), i + 1)
    plt.setp(ax, xticks=(), yticks=())

    polynomial_features = PolynomialFeatures(degree=degrees[i],
                                             include_bias=False)
    linear_regression = LinearRegression()
    pipeline = Pipeline([("polynomial_features", polynomial_features),
                         ("linear_regression", linear_regression)])
    pipeline.fit(X[:, np.newaxis], y)

    # Evaluate the models using crossvalidation
    scores = cross_val_score(pipeline, X[:, np.newaxis], y,
                             scoring="neg_mean_squared_error", cv=10)

    X_test = np.linspace(0, 1, 100)
    plt.plot(X_test, pipeline.predict(X_test[:, np.newaxis]), label="Model")
    plt.plot(X_test, true_fun(X_test), label="True function")
    plt.scatter(X, y, label="Samples")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.xlim((0, 1))
    plt.ylim((-2, 2))
    plt.legend(loc="best")
    plt.title("Degree {}\nMSE = {:.2e}(+/- {:.2e})".format(
        degrees[i], -scores.mean(), scores.std()))
plt.show()
```
